```python
# Create a data frame from the Public Settlement Report spreadsheet (this includes the cost for each site)

import pandas as pd

superfund_df = pd.read_excel('C:/Users/zthib/OneDrive/Desktop/EVE 310/Final Project Scripts/406192.xlsx', sheet_name = 'Region 6', header=15)
```

```python
# Display the df

superfund_df
```

```python
# Print column names for reference
print(superfund_df.columns)
```

```python
# Forward fill the site name-values extend across multiple rows
superfund_df['EPA_ID/NPL IND  TYPE'] = superfund_df['EPA_ID/NPL IND  TYPE'].ffill()
```

```python
# Display the df
superfund_df
```

```python
# Drop any rows without an 'amount' value
superfund_df = superfund_df.dropna(subset=['AMOUNT'])
```

```python
# Display the df
superfund_df
```

```python
# Create a new data frame with the site name/ID and amount
cost_df = cost_df = superfund_df[['EPA_ID/NPL IND  TYPE', 'AMOUNT']]
```

```python
# Display the df
cost_df
```

```python
# Group by site name
cost_df = cost_df.groupby('EPA_ID/NPL IND  TYPE')['AMOUNT'].sum().reset_index()
```

```python
# Display the df
cost_df
```

```python
# Extract only sites within Texas
tx_costs = cost_df[cost_df['EPA_ID/NPL IND  TYPE'].str.contains('TX')]
```

```python
# Display the df
tx_costs
```

```python
tx_costs.info()
```

```python
tx_costs.describe()
```

```python
import matplotlib.pyplot as plt
```

```python
# Create boxplot for cost distribution

fig1 = plt.figure()

plt.boxplot(tx_costs['AMOUNT'])
plt.title('Total Cost Distribution')
plt.ylabel('Total Cost ($)')

plt.show(fig1)
```

```python
# Create histogram for cost distribution

fig2 = plt.figure()

plt.hist(tx_costs['AMOUNT'])
plt.title('Total Cost Distribution')
plt.xlabel('Total Cost ($)')
plt.ylabel('Frequency')
```

```python
# Create histogram of only the first bin to show distribution without outliers

fig3 = plt.figure()

plt.hist(tx_costs['AMOUNT'], range=(0, 10_000_000), color='green')
plt.title('First Bin Cost Distribution')
plt.xlabel('Total Cost ($)')
plt.ylabel('Frequency')

plt.show()
```
