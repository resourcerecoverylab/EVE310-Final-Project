```python
# Import file with census tracts containing Superfund sites (created in ArcGIS)
superfund_df = pd.read_csv('C:/Users/zthib/OneDrive/Desktop/EVE 310/Final Project/superfund_tracts_w_id.csv')
```

```python
# Display the first five rows of the resulting data frame
superfund_df.head()
```

```python
# View basic information about the data frame
superfund_df.info()
```

```python
# Select only the FIPS column (this is the identifier for the census tract and is the only information necessary for the analysis)

superfund_tracts_df = superfund_df[['FIPS', 'EPA_ID']].copy()
```

```python
# Clean both data frames so the tract IDs are the same data type (string)
superfund_tracts_df['FIPS'] = superfund_tracts_df['FIPS'].astype(str).str.zfill(11)
census_data_df['LocationName'] = census_data_df['LocationName'].astype(str).str.zfill(11)
```

```python
superfund_cancer_df = pd.merge(
    census_data_df,               
    superfund_tracts_df[['FIPS', 'EPA_ID']],        
    left_on='LocationName',        
    right_on='FIPS',               
    how='inner'                    
)
```

```python
# View basic information about the data frame
superfund_cancer_df.info()
```

```python
# Drop one of the duplicate columns
superfund_cancer_df = superfund_cancer_df.drop(columns = ['LocationName'])
```
